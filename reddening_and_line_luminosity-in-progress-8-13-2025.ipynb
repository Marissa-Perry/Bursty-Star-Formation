## fixing datatype issue

```python
import ast  # for safely parsing strings into Python objects

# Safely convert stringified lists to real lists of floats
def parse_error_list(err_str):
    try:
        return ast.literal_eval(err_str)
    except (ValueError, SyntaxError):
        return [np.nan, np.nan, np.nan]  # or another fallback

# Apply to the column
NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1'] = NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1'].apply(parse_error_list)
NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 2'] = NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 2'].apply(parse_error_list)
NIRSpec_sample_df['NII INTEGRATED FLUX OBSERVED ERRORS 1'] = NIRSpec_sample_df['NII INTEGRATED FLUX OBSERVED ERRORS 1'].apply(parse_error_list)
NIRSpec_sample_df['NII INTEGRATED FLUX OBSERVED ERRORS 2'] = NIRSpec_sample_df['NII INTEGRATED FLUX OBSERVED ERRORS 2'].apply(parse_error_list)
NIRSpec_sample_df['H-BETA INTEGRATED FLUX OBSERVED ERRORS'] = NIRSpec_sample_df['H-BETA INTEGRATED FLUX OBSERVED ERRORS'].apply(parse_error_list)
```

# Correction for [NII] in PRISM sources

```python
# separating dataframe by data type
prism_df = NIRSpec_sample_df[NIRSpec_sample_df['SPECTRA TYPE'] == 'PRISM'].copy()
grating_df = NIRSpec_sample_df[NIRSpec_sample_df['SPECTRA TYPE'] == 'GRATING'].copy()
prism_grating_df = NIRSpec_sample_df[NIRSpec_sample_df['SPECTRA TYPE'] == 'PRISM AND GRATING'].copy()

# correction factor computed with grating line fluxes
Ha = grating_df['H-ALPHA INTEGRATED FLUX OBSERVED 1']
NII = grating_df['NII INTEGRATED FLUX OBSERVED 1']
frac = Ha / (Ha + NII)
corr_factor = np.median(frac)

# uncertainty on corr_factor
lower_err = corr_factor - np.percentile(frac, 16)
upper_err = np.percentile(frac, 84) - corr_factor
sigma_corr_factor = 0.5 * (lower_err + upper_err)

print('correction factor computed with grating line fluxes:',round(corr_factor,2),'±',round(sigma_corr_factor,2))

# correcting PRISM flux measurements ---------------------------------------
uncorr_flux = prism_df['H-ALPHA INTEGRATED FLUX OBSERVED 1']
uncorr_flux_err = prism_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1']
corr_flux = uncorr_flux * corr_factor
prism_df['H-ALPHA INTEGRATED FLUX OBSERVED 1'] = corr_flux

prism_corr_flux_errs = []
for i in range(len(prism_df)):
    flux = uncorr_flux.iloc[i]
    mean_err, err_16, err_84 = uncorr_flux_err.iloc[i]

    mean_err_corr = np.sqrt((corr_factor * mean_err)**2 + (flux * sigma_corr_factor)**2)
    err_16_corr   = np.sqrt((corr_factor * err_16)**2 + (flux * sigma_corr_factor)**2)
    err_84_corr   = np.sqrt((corr_factor * err_84)**2 + (flux * sigma_corr_factor)**2)

    prism_corr_flux_errs.append([mean_err_corr, err_16_corr, err_84_corr])

prism_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1'] = prism_corr_flux_errs
# ----------------------------------------------------------------------------

# updating the main dataframe
for col in prism_df.columns:
    if col in NIRSpec_sample_df.columns:
        NIRSpec_sample_df.loc[prism_df.index, col] = prism_df[col]
        
# correcting PRISM + Grating flux measurements -------------------------------
uncorr_flux = prism_grating_df['H-ALPHA INTEGRATED FLUX OBSERVED 1']
uncorr_flux_err = prism_grating_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1']
corr_flux = uncorr_flux * corr_factor
prism_grating_df['H-ALPHA INTEGRATED FLUX OBSERVED 1'] = corr_flux

prism_grating_corr_flux_errs = []
for i in range(len(prism_grating_df)):
    flux = uncorr_flux.iloc[i]
    mean_err, err_16, err_84 = uncorr_flux_err.iloc[i]

    mean_err_corr = np.sqrt((corr_factor * mean_err)**2 + (flux * sigma_corr_factor)**2)
    err_16_corr   = np.sqrt((corr_factor * err_16)**2 + (flux * sigma_corr_factor)**2)
    err_84_corr   = np.sqrt((corr_factor * err_84)**2 + (flux * sigma_corr_factor)**2)

    prism_grating_corr_flux_errs.append([mean_err_corr, err_16_corr, err_84_corr])

prism_grating_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1'] = prism_grating_corr_flux_errs
# ----------------------------------------------------------------------------

# updating the main dataframe
for col in prism_grating_df.columns:
    if col in NIRSpec_sample_df.columns:
        NIRSpec_sample_df.loc[prism_grating_df.index, col] = prism_grating_df[col]
```

```python
bad_Balmer_measurement = []
count_Ha_bad = 0
count_Hb_bad = 0
count_both_bad = 0
nan_values = 0
for source in range(len(NIRSpec_sample_df)):
    SNR_Ha = NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED 1'][source] / NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1'][source][0]
    SNR_Hb = NIRSpec_sample_df['H-BETA INTEGRATED FLUX OBSERVED'][source] / NIRSpec_sample_df['H-BETA INTEGRATED FLUX OBSERVED ERRORS'][source][0]
    
    if (SNR_Ha < 3) or (SNR_Hb < 3) or np.isnan(SNR_Ha) or np.isnan(SNR_Hb):
        bad_Balmer_measurement.append(NIRSpec_sample_df['MSA_ID'][source])   

    if (SNR_Ha < 3) and (SNR_Hb < 3):
        count_both_bad += 1
    
    elif SNR_Ha < 3:
        count_Ha_bad += 1

    elif SNR_Hb < 3:
        count_Hb_bad += 1

    if np.isnan(SNR_Ha) or np.isnan(SNR_Hb):
        nan_values += 1

print(str(nan_values)+' sources with nan values for either or both Balmer measurements')
print(str(count_both_bad) + ' sources with both Balmer measurements having SNR < 3')
print(str(count_Ha_bad)+' sources with Ha measurements having SNR < 3')
print(str(count_Hb_bad)+' sources with Hb measurements having SNR < 3')
print(str(len(bad_Balmer_measurement))+' TOTAL SOURCES to be dropped from dataframe')
```

# Reddening calculation

```python
Ha_Hb_intrinsic_lum_ratio = 2.87
Ha_Hb_extinction = 1.163
reddening_list = []
reddening_err_list = []

for source in range(len(NIRSpec_sample_df)):
    # always using first flux measurement for reddening calculation
    H1 = NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED 1'][source]
    H2 = NIRSpec_sample_df['H-BETA INTEGRATED FLUX OBSERVED'][source]
    
    # take the first value as symmetric 1-sigma error
    H1_err = NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1'][source][0]
    H2_err = NIRSpec_sample_df['H-BETA INTEGRATED FLUX OBSERVED ERRORS'][source][0]
    
    reddening, reddening_err = dust_reddening(H1, H1_err, H2, H2_err)

    reddening_list.append(reddening)
    reddening_err_list.append(reddening_err)  

NIRSpec_sample_df['REDDENING'] = reddening_list
NIRSpec_sample_df['REDDENING ERR'] = reddening_err_list

# Calzetti 2000 --> R_v = 4.05
NIRSpec_sample_df["A_V"] = 4.05 * NIRSpec_sample_df["REDDENING"]
NIRSpec_sample_df["A_V ERR"] = 4.05 * NIRSpec_sample_df["REDDENING ERR"]
```

# Dust correcting + obtaining integrated line flux & luminosity

```python
Ha_integ_lum = []
Ha_integ_lum_err = []
for source in range(len(NIRSpec_sample_df)):

    # need to dust correct 2nd measurement !
    if NIRSpec_sample_df['SPECTRA TYPE'][source] == 'PRISM AND GRATING':
        Ha_corr_flux, Ha_corr_flux_err = dust_correct_flux(NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED 2'][source], NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 2'][source], NIRSpec_sample_df['REDDENING'][source], NIRSpec_sample_df['REDDENING ERR'][source], 'Ha')

    # need to dust correct 1st measurement !
    else:
        Ha_corr_flux, Ha_corr_flux_err = dust_correct_flux(NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED 1'][source], NIRSpec_sample_df['H-ALPHA INTEGRATED FLUX OBSERVED ERRORS 1'][source], NIRSpec_sample_df['REDDENING'][source], NIRSpec_sample_df['REDDENING ERR'][source], 'Ha')

    
    Ha_lum = integrated_flux_to_luminosity(Ha_corr_flux,NIRSpec_sample_df['Z_SPEC'][source])
    Ha_lum_err = [integrated_flux_to_luminosity(Ha_corr_flux_err[0],NIRSpec_sample_df['Z_SPEC'][source]),
                   integrated_flux_to_luminosity(Ha_corr_flux_err[1],NIRSpec_sample_df['Z_SPEC'][source]),
                   integrated_flux_to_luminosity(Ha_corr_flux_err[2],NIRSpec_sample_df['Z_SPEC'][source])
                  ]

    Ha_integ_lum.append(Ha_lum)
    Ha_integ_lum_err.append(Ha_lum_err)
```

```python
# updating table
NIRSpec_sample_df.insert(6,'H-ALPHA INTEGRATED LUMINOSITY', Ha_integ_lum)
NIRSpec_sample_df.insert(7,'H-ALPHA INTEGRATED LUMINOSITY MEAN ERROR', [Ha_integ_lum_err[i][0] for i in range(len(NIRSpec_sample_df))])
NIRSpec_sample_df.insert(8,'H-ALPHA INTEGRATED LUMINOSITY 16% ERROR', [Ha_integ_lum_err[i][1] for i in range(len(NIRSpec_sample_df))])
NIRSpec_sample_df.insert(9,'H-ALPHA INTEGRATED LUMINOSITY 84% ERROR', [Ha_integ_lum_err[i][2] for i in range(len(NIRSpec_sample_df))])
NIRSpec_sample_df
```
